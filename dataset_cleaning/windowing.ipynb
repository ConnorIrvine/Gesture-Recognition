```python
# IMPORTS AND CONSTANTS

import math
import pandas as pd
import pickle
from pathlib import Path

WINDOW_SIZE = 0.2  # length of window in seconds
WINDOW_STEP = 0.15  # how much to slide the window
WORKSPACE = Path.cwd()
SRC_ROOT = WORKSPACE / 'osfstorage-archive'
DST_ROOT = WORKSPACE / 'pkl_dataset_resampled'
SUBJECTS = [f'subj{n:02d}' for n in range(1, 11)]
```

```python
# FUNCTIONS

def num_windows(length):
    ''' length: length of the EMG array '''
    return int(math.floor((length - WINDOW_SIZE) / WINDOW_STEP)) + 1


def load_pkl(file_path):
    """Load and return a Python object from a .pkl file."""
    p = Path(file_path)
    with p.open("rb") as f:
        return pickle.load(f)
```

```python
# LOAD EACH FILE OF THE DATASET

pkl_files = sorted(DST_ROOT.rglob("*.pkl"))
print("Found PKL files:", len(pkl_files))

all_trials = []
bad = []

for p in pkl_files:
    all_trials.append(load_pkl(p))
```
